# Data

Data is defined as a factual information.

Data can be stored in the digital form in computers and can be processed to get useful insights.

Data Science considers 4 types of data:

1. Numerical data or quantitative data

2. Categorical data or qualitative data

3. Ordinal data

4. Images, Audio and Video data

## Numerical or Quantitative Data

Numerical data is represented in the form of numbers and can be measured in quantities.

Numerical data is of 2 types:

1. Discrete Data

2. Continuous Data

### Discrete Data

Discrete data is represented as an integer or floating-point number.

Discrete data is useful to count something.

Example: The number of students in a class, the salary of an employee.

### Continuous Data

Continuos data moves continuosuly without ending. This can be upto infinity.

Example: Time, length, heigth

Time can be measured as hour, minutes, seconds, milliseconds, microseconds, nanoseconds and so on. Since we cannot measure this
continuous data due to limitation of out clocks, we restrict to hour, minute and seconds.

## Categorical or Qualitative Data

Categorical data has no inherent mathematical meaning.

Categorical data contains only a limited number of possibilities, which are called categories or classes.

Numbers are assigned to categories to represent them more compactly.

Example: 

Gender - Has only 2 possibilites, male and female.

Marital Status - Has possibilities like single, married, widowed, divorced.

Religion - Hindu, Christian, Muslim, Sikh, Jain, Buddhist.

## Ordinal Data

Ordinal data is a mixture of numerical and categorical data that has mathematical meaning.

Example:

Movie rating on a scale of 1 to 5. 1 means worse and 5 means excellent.

Size of a shirt: S/M/L/XL. S means small and XL means Extra-large.

## Images

An image contains several pixels (Picture element).

A pixel is a minute dot with which a character or image is composed of.

These pixels on the computer monitor are not visible to the nacked eye since they are all merged.

The pixels can be represented by either 1 or 0.

A bright pixel is represented by a 1 and a dark pixel is represented by a 0.

An image can be converted into a matrix of 1s and os.

A color image has 3 colors of pixels: Red (R), Green (G) and Blue (B), called primary colors.

All colors in the universe are combination of the primary colors.

Each pixel in the colour image is represented in the form of 3 pixels.

## Audio

An audio file indicates variations in the intensity of sound.

The variations are represented by the values from 0 to 1.

## Video Data

A video file indicates a group of images, called frames, that are moved quickly such that our eyes cannot identify them separately.

The images are displayed quickly such that they give the illusion that they are moving.

A video contains several hundreds of images which all can be converted into 1s and 0s.

## Types of Machine Learning

The main task of machine learning is to explore and construct algorithms that can learn from historical data and make predictions on new input data.

Depending on the nature of the learning data, machine learning tasks can be broadly classified into the following three categories:

1. Supervised Learning:

When learning data comes with a description, targets, or desired output besides indicative signals (called as labeled data), the learning goal is to find a general rule that maps input to output.

The learned rule is then used to label new data with unknown output.

Example: Face and speech recognition, product or movie recommendations, sales forecasting, spam email detection.

2. Unsupervised Learning:

When the learning data only contains indicative signals without any description attached (called as unlabeled data),  we need to find the structure of the data underneath, discover hidden information, or determine how to describe the data.

Example: Anomalies detection, group customers with similar online behaviors for a marketing campaign.

3. Reinforcement learning:

Learning data provides feedback so that a system adapts to dynamic conditions in order to ultimately achieve a certain goal. The system evaluates its performance based on the feedback responses and reacts accordingly.

Example: Robotics for industrial automation, self-driving cars, and the chess master AlphaGo.



Supervised Learning can further subdivided into regression and classification.

**Regression** trains on and predicts continuous-valued responses. Example: Predicting house prices.

**Classification** attempts to find the appropriate class label. Example: Analyzing a positive/negative sentiment, predicting a loan default.

**Semi-supervised** learning makes use of unlabeled data (typically a large amount) for training, besides a small amount of labeled data.

## Generalizing with data

The world generates lot of data and it's hard to process this data.

Humans receive the data through ears and eyes. These inputs are transformed into electrical or chemical signals.

Computers and robots also work wwith electrical signals.

The electrical signals are translated into 1s and 0s.

While programming the data are represented either as numbers or texts. Text is not very convenient and need to transform into numerical values.

The supervised learning has a scenario similar to studying for an exam: Has a set of practice questions and the actual exam.

**Generalization** - Learn from practice questions and apply this knowledge to answer other similar questions in the exam.

In machine learning, the practice questions are called training sets or training samples.

The machine learning models derive patterns from training sets.

The actual exams are the testing sets or testing samples.

The testing samples are eventually applied to the machine learning models.

Learning effectiveness is measured by the compatibility of the learning models and the testing.

Sometimes, between practice questions and actual exams, there will be a mock exams, known as validation sets or validation samples in machine learning.

Validation samples help to verify how well the models will perform in a simulated setting, and then model is fine-tuned in order to achieve greater accuracy.

In supervised learning the predicted results are compared against the expected values.
